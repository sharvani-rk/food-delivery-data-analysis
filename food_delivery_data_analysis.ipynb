```python
import pandas as pd
import sqlite3
```

```python
## Datasets Used
#- orders.csv → Transactional data
#- users.json → User master data
#- restaurants.sql → Restaurant master data
```

```python
orders = pd.read_csv("orders.csv")
users = pd.read_json("users.json")
conn = sqlite3.connect(":memory:")
with open("restaurants.sql", "r") as f:
    sql_script = f.read()
conn.executescript(sql_script)
restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
```

```python
orders_users = pd.merge(
    orders,
    users,
    on="user_id",
    how="left"
)
final_df = pd.merge(
    orders_users,
    restaurants,
    on="restaurant_id",
    how="left"
)
final_df.head()
```

```python
## Key Insights
#- Gold members contribute a significant portion of revenue
#- Certain cuisines have higher average order values
#- High-rated restaurants generate more revenue
```
